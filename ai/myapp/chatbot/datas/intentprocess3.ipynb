# 제품_재고/제품_정보/제품_추천/제품_일반/제품_기타/ 결제(배송,결제,구매) / 매장(웹사이트, 매장) / 기타(예약,부가서비스,주문, 포장, 멤버십, 행사) / AS(교환|환불|반품, AS)


```python
# Load the provided final_df.csv
import pandas as pd

final_df = pd.read_csv("./query_and_intent_only_drop_dup_first_try.csv")

def labeling(df):
    # Define a mapping based on the given classification
    labels = {
        '제품_재고': 0, '제품_정보': 1, '제품_추천': 2, '제품_일반': 3,
        '제품_기타': 4, '결제': 5, '매장': 6, '기타': 7, 'AS': 8
    }
    
    label_mapping = {
        '제품_재고': '제품_재고',
        '제품_정보': '제품_정보',
        '제품_추천': '제품_추천',
        '제품_일반': '제품_일반',
        '배송': '결제',
        '결제': '결제',
        '구매': '결제',
        '웹사이트': '매장',
        '매장': '매장',
        '예약': '기타',
        '부가서비스': '기타',
        '주문': '기타',
        '포장': '기타',
        '멤버십': '기타',
        '행사': '기타',
        '교환|반품|환불': 'AS',
        'AS': 'AS'
    }
    
    def get_label(intent):
        main_intent = intent.split('_')[0]
        sub_intent = intent.split('_')[1] if '_' in intent else None
        
        # For '제품' main intent
        if main_intent == '제품':
            combined_intent = f"{main_intent}_{sub_intent}" if sub_intent else main_intent
            return labels.get(label_mapping.get(combined_intent, '제품_기타'))
        else:
            return labels.get(label_mapping.get(main_intent, '기타'))
    
    # Apply the get_label function to the 'intent' column
    df['label'] = df['intent'].apply(get_label)
    return df

# Apply the labeling function to the dataframe
labeled_df = labeling(final_df)
label_counts = labeled_df['label'].value_counts().sort_index()

label_counts
```
