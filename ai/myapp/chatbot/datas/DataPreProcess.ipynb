```python
import pandas as pd
```

```python
df = pd.read_csv('./clothes_train.csv')
```

```python
df.head(5)
```

```python
new_df = df[['IDX', '발화자', '발화문', '감성', '인텐트',]]
n=new_df
n
```

```python
intents = df['인텐트'].unique()
print(intents)
```

```python
n2 = n[n['발화자'] == 'c']

n2
n3=n2[['발화문', '감성', '인텐트']]
n3
```

```python
# 발화문과 인텐트를 빼내어 새로운 데이터프레임을 생성합니다.
df_intent = n3[['발화문', '인텐트']].copy()
df_intent.columns = ['query', 'intent']

# 생성된 데이터프레임을 intent.csv로 저장합니다.
df_intent.to_csv('intent3.csv', index=False)

# 발화문과 감성을 빼내어 새로운 데이터프레임을 생성합니다.
df_sense = df[['발화문', '감성']].copy()
df_sense.columns = ['query', 'sense']

# 생성된 데이터프레임을 sense.csv로 저장합니다.
df_sense.to_csv('sense.csv', index=False)
```
